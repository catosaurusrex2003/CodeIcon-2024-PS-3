```python


# print(tables[0].df.iloc[0])
to_del = []
for i in range(len(tables)):
    case = False
    k = 0
    for j in range(len(tables.iloc[i])):
        # tables.loc[i,j] = tables.loc[i+1,0].split()
        if tables.iloc[i,j] not in [' ','',"R’m"]:
            k+=1
        # print(i,tables.iloc[i,j])
    if k<3:
        to_del.append(i)
# print(to_del)
for i in to_del:
    # print(i)
    if i+1<len(tables.iloc[i]):
        tables.loc[i+1,0] = tables.loc[i,0] + ' ' + tables.loc[i+1,0]
tables = tables.drop(labels=to_del,axis=0)
tables
```
